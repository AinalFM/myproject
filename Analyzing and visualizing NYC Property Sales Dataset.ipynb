# 7. Confidence Intervals

Dalam statistika, confidence interval (interval kepercayaan) adalah kisaran nilai yang menggambarkan seberapa yakin kita bahwa parameter populasi terletak dalam interval tersebut. Ketika bekerja dengan distribusi t (t-distribution), langkah-langkah berikut dapat membantu membaca hasil confidence intervals:

Memahami distribusi t: Distribusi t adalah distribusi probabilitas yang sering digunakan ketika sampel yang digunakan dalam analisis kita kecil atau ketika variasi populasi tidak diketahui. Distribusi ini memiliki bentuk seperti lonceng (bell-shaped) dengan ekor yang lebih lebar dibandingkan distribusi normal.

Memahami interval kepercayaan: Confidence interval adalah kisaran nilai yang mengandung nilai yang kita yakini mengandung parameter populasi dengan tingkat kepercayaan tertentu. Misalnya, jika kita memiliki confidence interval 90%, itu berarti kita yakin 90% bahwa parameter populasi berada dalam interval tersebut.

Mengidentifikasi hasil confidence interval: Biasanya, hasil confidence interval ditampilkan dalam bentuk dua angka, yaitu lower bound (batas bawah) dan upper bound (batas atas). Misalnya, jika hasil confidence interval adalah (0.2, 0.8), ini berarti bahwa interval kepercayaan tersebut adalah 0.2 hingga 0.8.

Menggunakan derajat kebebasan (degrees of freedom): Dalam distribusi t, derajat kebebasan adalah parameter yang digunakan untuk menghitung interval kepercayaan. Derajat kebebasan umumnya tergantung pada ukuran sampel yang digunakan dalam analisis kita. Semakin besar ukuran sampel, semakin dekat distribusi t dengan distribusi normal. Untuk membaca hasil interval kepercayaan, kita perlu mengetahui derajat kebebasan yang digunakan.

```python
# menghitung derajat kebebasan 
n_data = len(df_new)
degree_freedom = n_data -1

print(degree_freedom)
```

```python
# create 90% confidence interval
st.t.interval(alpha=0.90, df=len(df_new)-1,
              loc=np.mean(df_new),
              scale=st.sem(df_new))
```

# 8. Hypothesis Testing

```python
df_new.head()
```

```python
df['TAX_CLASS_AT_TIME_OF_SALE'].unique()
```

Uji apakah rata-rata SALE PRICE untuk tingkat pajak properti unik (Tax 4) lebih mahal dari rata-rata SALE PRICE untuk tingkat pajak banyak properti (Tax 2).

h0 = rata-rata SALE PRICE untuk Tax 4 tidak lebih tinggi daripada Tax 2
h1 = rata-rata SALE PRICE untuk Tax 4 lebih tinggi daripada Tax 2
testing

h0 : mean_tax4 <= mean_tax2
h1 : mean_tax4 > mean_tax2

```python
tax4 = df_new[df_new['TAX_CLASS_AT_TIME_OF_SALE']==4]
tax2 = df_new[df_new['TAX_CLASS_AT_TIME_OF_SALE']==2]
```

```python
n_tax4 = len(tax4)
mean_tax4 = tax4['SALE_PRICE'].astype('float64').mean()
std_tax4 = tax4['SALE_PRICE'].astype('float64').std()
print(n_tax4,mean_tax4,std_tax4)
```

```python
n_tax2 = len(tax2)
mean_tax2 = tax2['SALE_PRICE'].astype('float64').mean()
std_tax2 = tax2['SALE_PRICE'].astype('float64').std()
print(n_tax2,mean_tax2,std_tax2)
```

```python
# menggunakan confidence level sebesar 95%
alpha = 0.05
```

```python
hipo_sale = sm.stats.ztest(tax4['SALE_PRICE'].astype('float64').dropna(),tax2['SALE_PRICE'].astype('float64').dropna(),alternative='larger')
hipo_sale
```

```python
if hipo_sale[1]>alpha:
    print('Terima h0 : rata-rata SALE PRICE untuk Tax 4 tidak lebih tinggi daripada Tax 2')
else:
    print('Tolak h0 : rata-rata SALE PRICE untuk Tax 4 lebih tinggi daripada Tax 2')
```

Berdasarkan hasil hipotesis tersebut dapat disimpulkan bahwa rata-rata harga jual properti untuk tingkat pajak properti unik (Tax 4) lebih tinggi daripada tingkat pajak banyak properti (Tax 2). Kesimpulan lainnya yaitu Tingkat pajak properti yang dikenakan pada saat penjualan (TAX CLASS AT TIME OF SALE) dapat mempengaruhi harga jual properti (SALE PRICE) karena pajak properti yang lebih tinggi dapat mempengaruhi harga yang diminta oleh penjual.
